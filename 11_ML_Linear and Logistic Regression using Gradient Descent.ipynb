```python
# data X needs ones in the first column
X = np.insert(X, 0, 1, axis=1) # X 데이터의 첫 번째 열에 1로 이루어진 열을 추가하여
X                              #  행렬 연산을 위한 상수항을 포함시킵니다.
```

```python
# make linear function: h(x)
# 선형 함수 h(x)를 정의합니다. theta는 가중치(weight) 벡터이고, x는 입력 변수입니다.
def linear_function(theta, x):
  '''
  theta: w0, w1, ..., wn
  x: x0=1, x1, ..., xn
  '''
  return np.dot(theta, x)
```

```python
X[1] # 1번 행
```

```python
# test linear function
# 비용 함수 J(theta)를 정의합니다.
theta_test = np.random.rand(4)
linear_function(theta_test, X[1])
```

```python
# define cost function
# 이 함수는 예측값과 실제값 사이의 오차의 제곱을 계산하고, 평균을 구합니다.
def cost_linear(theta, X, y):
  '''
  theta: w0, w1, ..., wn
  X: data
  y: predictor variable
  '''
  # m: number of data
  m = len(X)

  cost = 0
  for i in range(m):
    cost += np.power(linear_function(theta, X[i]) - y[i], 2)
  cost = cost * 1/m
  return cost

# 이 비용은 모델의 예측 성능을 평가하고, 경사 하강법에서 최적화할 목적 함수로 사용됩니다.

# theta는 선형 회귀 모델의 가중치(계수)를 나타내며,
# theta = [w0, w1, ..., wn] 형태의 벡터입니다.
# 이 가중치는 입력 변수(X)와 함께 선형 함수를 통해 예측값을 계산하는 데 사용
# 경사 하강법은 이 가중치를 최적화하여 오차를 최소화하는 값을 찾는 알고리즘입니다.
```

위 코드는 선형 회귀 모델의 비용 함수 J(theta)를 계산하는 함수입니다. 비용 함수는 모델의 예측값과 실제값 사이의 오차를 측정하여 모델의 성능을 평가하는 데 사용됩니다.

함수의 입력은 다음과 같습니다:

- `theta`: 가중치(w0, w1, ..., wn)를 포함한 벡터
- `X`: 입력 데이터
- `y`: 실제값 (예측 변수)

함수 내부에서는 다음과 같은 작업이 수행됩니다:

1. `m = len(X)`: 데이터의 개수를 m에 저장합니다.
2. `cost = 0`: 비용 값을 초기화합니다.
3. `for i in range(m):`: 데이터 개수만큼 반복합니다.
    - `cost += np.power(linear_function(theta, X[i]) - y[i], 2)`: 예측값과 실제값의 차이의 제곱을 누적하여 비용을 계산합니다.
4. `cost = cost * 1/m`: 누적된 비용을 데이터 개수로 나누어 평균 비용을 구합니다.

마지막으로 `cost` 값을 반환하여 비용 함수의 결과를 돌려줍니다. 이 비용은 모델의 예측 성능을 평가하고, 경사 하강법에서 최적화할 목적 함수로 사용됩니다.

theta는 선형 회귀 모델의 가중치(계수)를 나타내며, theta = [w0, w1, ..., wn] 형태의 벡터입니다. 이 가중치는 입력 변수(X)와 함께 선형 함수를 통해 예측값을 계산하는 데 사용됩니다. 경사 하강법은 이 가중치를 최적화하여 오차를 최소화하는 값을 찾는 알고리즘입니다.

```python
# test cost function
cost_linear(theta_test, X, y)
```

```python
# 비용 함수 J(theta)의 각 가중치에 대한 편미분인 기울기를 계산하는 함수입니다.
# j는 특정 가중치의 인덱스를 의미하며,
# dJ_dw는 해당 가중치에 대한 기울기 값을 반환합니다.
def compute_jth_linear_gradient(theta, j, X, y):
  '''
  theta: w0, w1, ..., wn
  j: jth feature
  X: data
  y: predictor variable
  '''
  # m: number of data
  m = len(X)

  dJ_dw = 0
  for i in range(m):
    dJ_dw += (linear_function(theta, X[i]) - y[i]) * X[i][j]
  dJ_dw = dJ_dw * (2/m)

  return dJ_dw
```

위 코드는 선형 회귀 모델의 비용 함수 J(theta)를 편미분하여 각 가중치(w0, w1, ..., wn)에 대한 기울기를 계산하는 함수입니다.

함수의 입력은 다음과 같습니다:

- `theta`: 가중치(w0, w1, ..., wn)를 포함한 벡터
- `j`: 특정 가중치의 인덱스
- `X`: 입력 데이터
- `y`: 실제값 (예측 변수)

함수 내부에서는 다음과 같은 작업이 수행됩니다:

1. `m = len(X)`: 데이터의 개수를 m에 저장합니다.
2. `dJ_dw = 0`: 기울기 값을 초기화합니다.
3. `for i in range(m):`: 데이터 개수만큼 반복합니다.
    - `dJ_dw += (linear_function(theta, X[i]) - y[i]) * X[i][j]`: 오차 (예측값과 실제값의 차이)에 입력 데이터의 j번째 특성 값을 곱하여 기울기를 누적합니다.
4. `dJ_dw = dJ_dw * (2/m)`: 누적된 기울기 값을 데이터 개수로 나누어 기울기의 평균을 구합니다.

마지막으로 `dJ_dw` 값을 반환하여 특정 가중치에 대한 기울기 값을 계산합니다. 이러한 기울기 값은 경사 하강법에서 가중치를 업데이트하는 데 사용됩니다.

```python
# test gradient
compute_jth_linear_gradient(theta_test, 1, X, y)
```

```python
# 경사하강법
def linear_gradient_descent(X, y):
  '''
  X: data
  y: predictor variable
  '''
  # m: number of data, n: number of feature
  m = len(X)
  n = X.shape[1] # 특성(열) 개수 저장

  # hyperparameter
  alpha = 0.01 # learning rate
  epsilon = 10 ** -4 # 수렴조건 (비용의 차이가 이 값보다 작으면 알고리즘 중지)

  # initialization
  theta = np.random.rand(n) # 가중치 무작위로 초기화

  # cost J
  J = np.zeros(1000) # 비용함수 J(theta) 저장 배열 초기화

  # gradient descent
  iter = 1 # 반복 횟수 초기화
  while True:

    # update gradient
    for j in range(n):
      theta[j] = theta[j] - alpha * compute_jth_linear_gradient(theta, j, X, y)
      #  해당 가중치를 "학습률"과 "해당 가중치에 대한 기울기"를 곱한 값으로 업데이트

    # compute cost
    J[iter] = cost_linear(theta, X, y) # 현재 가중치로 계산한 비용을 J 배열에 저장
    print(f"cost at time {iter}: {J[iter]}") # 현재 반복에서의 비용을 출력
    print(f"cost difference at {iter}: {np.abs(J[iter] - J[iter - 1])}") # 비용의 변화 출력

    # convergence condition
    # 비용의 변화가 수렴 조건을 만족하면 반복 중지
    if np.abs(J[iter] - J[iter - 1]) < epsilon:
      break

    iter += 1 # 반복 횟수를 증가

  return
# 최종적으로 함수는 반환값이 없고,
# 최적화된 가중치 theta를 사용하여 모델을 학습시킵니다.
# 이러한 과정을 통해 경사 하강법을 사용하여 선형 회귀 모델의 가중치를 최적화하고,
# 비용을 최소화하는 방향으로 모델을 업데이트합니다.
```

위 코드는 경사 하강법을 사용하여 선형 회귀 모델의 가중치를 최적화하는 함수입니다.

함수의 입력은 다음과 같습니다:
- `X`: 입력 데이터
- `y`: 실제값 (예측 변수)

함수 내부에서는 다음과 같은 작업이 수행됩니다:

1. `m = len(X)`: 데이터의 개수를 m에 저장합니다.
2. `n = X.shape[1]`: 입력 데이터의 특성(열) 개수를 n에 저장합니다.
3. `alpha = 0.01`: 학습률(learning rate)로, 가중치를 업데이트할 때 사용되는 스텝의 크기를 결정합니다.
4. `epsilon = 10 ** -4`: 수렴 조건으로, 비용의 차이가 이 값보다 작으면 알고리즘을 중지합니다.
5. `theta = np.random.rand(n)`: 가중치를 무작위로 초기화합니다.
6. `J = np.zeros(1000)`: 비용 함수 J(theta)의 값들을 저장할 배열을 초기화합니다.
7. `iter = 1`: 반복 횟수를 초기화합니다.

그리고 아래의 작업들을 반복하면서 경사 하강법을 수행합니다:

8. `for j in range(n):`: 각 가중치에 대해 반복합니다.
   - `theta[j] = theta[j] - alpha * compute_jth_linear_gradient(theta, j, X, y)`: 해당 가중치를 학습률과 해당 가중치에 대한 기울기를 곱한 값으로 업데이트합니다.
9. `J[iter] = cost_linear(theta, X, y)`: 현재 가중치로 계산한 비용을 J 배열에 저장합니다.
10. `print(f"cost at time {iter}: {J[iter]}")`: 현재 반복에서의 비용을 출력합니다.
11. `print(f"cost difference at {iter}: {np.abs(J[iter] - J[iter - 1])}")`: 현재 반복에서의 비용의 변화를 출력합니다.
12. `if np.abs(J[iter] - J[iter - 1]) < epsilon: break`: 비용의 변화가 수렴 조건을 만족하면 반복을 중지합니다.
13. `iter += 1`: 반복 횟수를 증가시킵니다.

최종적으로 함수는 반환값이 없고, 최적화된 가중치 theta를 사용하여 모델을 학습시킵니다.

이러한 과정을 통해 경사 하강법을 사용하여 선형 회귀 모델의 가중치를 최적화하고, 비용을 최소화하는 방향으로 모델을 업데이트합니다.

```python
# test linear regression
result = linear_gradient_descent(X, y)
```
